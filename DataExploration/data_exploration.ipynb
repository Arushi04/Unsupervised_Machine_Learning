### Q2- Display summary statistics of the data.  What can you learn from it on the data?

```python
housing_data.describe()
```

**Observations:** 
1. Count gives the count of all the features. We see that the total bedrooms are shown as 20433 which means the summary is automatically ignoring the null values.
2. Min and Max are simply giving the min and max values of the features. And mean is giving the mean value of all numerical features.    
3. The summary stats are only for numerical features. So ocean_proximity is not included here.
4. Std is standard daviation and it shows how dispersed the values are.
5. 25%, 50%, and 75% rows show the corresponding percentiles. A percentile indicates the value below which a given percentage of observations in a group of observations fall. 
For eg, 25% of the districts have a housing_median_age lower than 18, while 50% are lower than 29 and 75% are lower than 37.
